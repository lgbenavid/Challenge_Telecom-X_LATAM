```python
print("\n=== OUTLIERS EN VARIABLES NUMRICAS ===")
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

for col in numeric_cols:
    if df[col].nunique() > 5:  # Solo para variables con suficiente variaci贸n
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_percent = (len(outliers) / len(df)) * 100

        if len(outliers) > 0:
            print(f"\nColumna: {col}")
            print(f"Outliers detectados: {len(outliers)} ({outlier_percent:.2f}%)")
            print(f"Rango normal: [{lower_bound:.2f}, {upper_bound:.2f}]")

            # Visualizaci贸n
            plt.figure(figsize=(8, 4))
            plt.boxplot(df[col].dropna(), vert=False)
            plt.title(f'Boxplot de {col}')
            plt.show()
```

```python
print("\n=== CORRECCIN DE INCONSISTENCIAS LGICAS ===")

# 1. Corregir TotalCharges para tenure=0
if 'TotalCharges' in df_clean.columns and 'tenure' in df_clean.columns:
    mask = (df_clean['tenure'] == 0) & (df_clean['TotalCharges'] > 0)
    df_clean.loc[mask, 'TotalCharges'] = df_clean.loc[mask, 'MonthlyCharges']
    print(f"Corregidos {mask.sum()} casos de TotalCharges inconsistentes")

# 2. Corregir servicios no contratados pero con caracter铆sticas
if 'PhoneService' in df_clean.columns and 'MultipleLines' in df_clean.columns:
    mask = (df_clean['PhoneService'] == 'No') & (df_clean['MultipleLines'] != 'No Phone Service')
    df_clean.loc[mask, 'MultipleLines'] = 'No Phone Service'
    print(f"Corregidos {mask.sum()} casos de MultipleLines inconsistentes")

# 3. Asegurar que los servicios adicionales sean "No Internet Service" cuando no hay Internet
internet_dependent_cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                          'TechSupport', 'StreamingTV', 'StreamingMovies']

if 'InternetService' in df_clean.columns:
    for col in internet_dependent_cols:
        if col in df_clean.columns:
            mask = (df_clean['InternetService'] == 'No') & (~df_clean[col].isin(['No Internet Service', 'No']))
            df_clean.loc[mask, col] = 'No Internet Service'
            print(f"Corregidos {mask.sum()} casos en {col}")
```

## **16. Manejo de Outliers**

```python
print("\n=== MANEJO DE OUTLIERS ===")

# Estrategia conservadora: marcar outliers pero no eliminarlos
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
outliers_report = []

for col in numeric_cols:
    if df_clean[col].nunique() > 5:  # Solo para variables con suficiente variaci贸n
        q1 = df_clean[col].quantile(0.25)
        q3 = df_clean[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        outliers = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)]
        outlier_percent = (len(outliers) / len(df_clean)) * 100

        if len(outliers) > 0:
            # Crear columna indicadora de outliers
            df_clean[f'{col}_outlier'] = ((df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)).astype(int)
            outliers_report.append({
                'Columna': col,
                'Outliers': len(outliers),
                'Porcentaje': outlier_percent,
                'L铆mite inferior': lower_bound,
                'L铆mite superior': upper_bound
            })

if outliers_report:
    print("\nOutliers identificados (se han creado columnas indicadoras):")
    print(pd.DataFrame(outliers_report))
else:
    print("No se identificaron outliers significativos")
```
