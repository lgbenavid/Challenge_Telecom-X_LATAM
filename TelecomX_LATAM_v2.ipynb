```python
# Análisis de variables categóricas

# Seleccionar variables categóricas
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

if 'Churn' in df.columns and len(categorical_cols) > 0:
    print("\n=== ANÁLISIS DE VARIABLES CATEGÓRICAS ===")

    # Eliminar columnas no relevantes
    categorical_cols = [col for col in categorical_cols if col not in ['customerID', 'Churn']]

    print("\nTasa de Churn por categoría:")

    # Filtrar columnas que contienen tipos no hashables como dict o list
    processable_categorical_cols = []
    for col in categorical_cols:
        # Comprueba el tipo del primer valor no nulo
        first_non_null = df[col].dropna().iloc[0] if not df[col].dropna().empty else None
        if first_non_null is not None and not isinstance(first_non_null, (dict, list)):
            processable_categorical_cols.append(col)
        else:
            print(f"Excluyendo columna '{col}' del análisis categórico debido a tipos de datos complejos (dict/list).")

    for col in processable_categorical_cols:
        if df[col].nunique() < 10:  # Evitar columnas con demasiadas categorías
            # Asegúrese que 'Churn_binary' exista antes de calcular churn_rate
            if 'Churn_binary' in df.columns:
                churn_rate = df.groupby(col)['Churn_binary'].mean().sort_values(ascending=False)
                # Asegúrese que churn_rate no esté vacío antes de trazar el gráfico
                if not churn_rate.empty:

                    # Visualización
                    import matplotlib.pyplot as plt # Importar matplotlib
                    churn_rate.plot(kind='bar', title=f'Tasa de Churn por {col}')
                    plt.show()
                else:
                    print(f"No hay datos para calcular la tasa de Churn para '{col}'.")
            else:
                print("Error: La columna 'Churn_binary' no existe. Asegúrese de haberla creado.")
else:
    print("\nError: No se encontró la columna 'Churn' o no hay columnas categóricas para analizar.")
```

```python
# Identificación de variables más relevantes

# Variables típicamente más relevantes para Churn en telecomunicaciones
important_features = {
    'Contrato': ['tenure', 'Contract', 'PaperlessBilling'],
    'Servicio': ['InternetService', 'OnlineSecurity', 'TechSupport'],
    'Pagos': ['PaymentMethod', 'MonthlyCharges', 'TotalCharges'],
    'Demográficas': ['SeniorCitizen', 'Partner', 'Dependents']
}

print("\n=== VARIABLES CLAVE PARA PREDICCIÓN DE CHURN ===")
for category, features in important_features.items():
    available_features = [f for f in features if f in df.columns]
    if available_features:
        print(f"\n{category}:")
        print(", ".join(available_features))

        # Defina numeric_cols y categorical_cols
        # Asegurarse de que las variables 'numeric_cols' y 'categorical_cols' estén definidas
        numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
        # También necesitamos 'categorical_cols' para la otra parte del if/elif
        categorical_cols = df.select_dtypes(include=['object']).columns.tolist()


        # Mostrar impacto en Churn para cada característica disponible
        for feature in available_features:
            if feature in numeric_cols:
                # Asegúrese de que 'Churn_binary' exista antes de calcular la correlación
                if 'Churn_binary' in df.columns:
                    corr = df[['Churn_binary', feature]].corr().iloc[0,1]
                    print(f"  - {feature}: Correlación = {corr:.2f}")
                else:Comprueba si la serie resultante está vacía
                    print(f"  - {feature}: No se pudo calcular correlación, 'Churn_binary' no existe.")
            elif feature in categorical_cols:
                # Asegúrese de que exista 'Churn_binary' antes de calcular la tasa de abandono media
                if 'Churn_binary' in df.columns:
                    if df[feature].nunique() < 10:  # Comprueba si hay suficientes valores únicos
                        # Comprueba si hay valores no nulos en la función y Churn_binary
                        # Esta es una forma más confiable de comprobar si la agrupación producirá datos.
                        if not df[df[feature].notna() & df['Churn_binary'].notna()].empty:
                            churn_rate = df.groupby(feature)['Churn_binary'].mean().sort_values(ascending=False)
                            # Evite realizar cálculos en grupos vacíos, lo que podría causar errores después de la agrupación.
                            # Check if the resulting Series is empty
                            if not churn_rate.empty:
                                max_churn = churn_rate.max()
                                min_churn = churn_rate.min()
                                print(f"  - {feature}: Ratio Churn varía entre {min_churn*100:.1f}% y {max_churn*100:.1f}%")
                            else:
                                print(f"  - {feature}: No hay datos para calcular la tasa de Churn para '{feature}'.")
                        else:
                            print(f"  - {feature}: No hay datos suficientes para calcular el Ratio Churn.")
                    else:
                         print(f"  - {feature}: Demasiadas categorías para un análisis simple del Ratio Churn.")
                else:
                    print(f"  - {feature}: No se pudo calcular Ratio Churn, 'Churn_binary' no existe.")
```

## **8. Análisis de outliers en variables numéricas**

```python
print("\n=== OUTLIERS EN VARIABLES NUMÉRICAS ===")
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

for col in numeric_cols:
    if df[col].nunique() > 5:  # Solo para variables con suficiente variación
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_percent = (len(outliers) / len(df)) * 100

        if len(outliers) > 0:
            print(f"\nColumna: {col}")
            print(f"Outliers detectados: {len(outliers)} ({outlier_percent:.2f}%)")
            print(f"Rango normal: [{lower_bound:.2f}, {upper_bound:.2f}]")

            # Visualización
            plt.figure(figsize=(8, 4))
            plt.boxplot(df[col].dropna(), vert=False)
            plt.title(f'Boxplot de {col}')
            plt.show()
```

```python
print("\n=== ESTANDARIZACIÓN DE CATEGÓRICAS ===")

# Diccionario de estandarización
standardization_map = {
    'gender': {'Male': 'Male', 'Female': 'Female', 'M': 'Male', 'F': 'Female'},
    'Partner': {'Yes': 'Yes', 'No': 'No', 'Y': 'Yes', 'N': 'No'},
    'Dependents': {'Yes': 'Yes', 'No': 'No', 'Y': 'Yes', 'N': 'No'},
    'Churn': {'Yes': 'Yes', 'No': 'No', 'Churned': 'Yes', 'Stayed': 'No'}
}

for col, mapping in standardization_map.items():
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].map(mapping).fillna(df_clean[col])
        print(f"Estandarizados valores en {col}")

# Estandarización de texto (primera letra mayúscula)
text_cols = [col for col in categorical_cols if col not in ['customerID']]
for col in text_cols:
    df_clean[col] = df_clean[col].str.title()

print("\nValores categóricos estandarizados:")
for col in text_cols[:3]:  # Mostrar solo algunas columnas como ejemplo
    print(f"{col}: {df_clean[col].unique()}")
```

## **15. Corrección de Inconsistencias Lógicas**

```python
print("\n=== CORRECCIÓN DE INCONSISTENCIAS LÓGICAS ===")

# 1. Corregir TotalCharges para tenure=0
if 'TotalCharges' in df_clean.columns and 'tenure' in df_clean.columns:
    mask = (df_clean['tenure'] == 0) & (df_clean['TotalCharges'] > 0)
    df_clean.loc[mask, 'TotalCharges'] = df_clean.loc[mask, 'MonthlyCharges']
    print(f"Corregidos {mask.sum()} casos de TotalCharges inconsistentes")

# 2. Corregir servicios no contratados pero con características
if 'PhoneService' in df_clean.columns and 'MultipleLines' in df_clean.columns:
    mask = (df_clean['PhoneService'] == 'No') & (df_clean['MultipleLines'] != 'No Phone Service')
    df_clean.loc[mask, 'MultipleLines'] = 'No Phone Service'
    print(f"Corregidos {mask.sum()} casos de MultipleLines inconsistentes")

# 3. Asegurar que los servicios adicionales sean "No Internet Service" cuando no hay Internet
internet_dependent_cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                          'TechSupport', 'StreamingTV', 'StreamingMovies']

if 'InternetService' in df_clean.columns:
    for col in internet_dependent_cols:
        if col in df_clean.columns:
            mask = (df_clean['InternetService'] == 'No') & (~df_clean[col].isin(['No Internet Service', 'No']))
            df_clean.loc[mask, col] = 'No Internet Service'
            print(f"Corregidos {mask.sum()} casos en {col}")
```

## **16. Manejo de Outliers**

```python
print("\n=== MANEJO DE OUTLIERS ===")

# Estrategia conservadora: marcar outliers pero no eliminarlos
numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
outliers_report = []

for col in numeric_cols:
    if df_clean[col].nunique() > 5:  # Solo para variables con suficiente variación
        q1 = df_clean[col].quantile(0.25)
        q3 = df_clean[col].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        outliers = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)]
        outlier_percent = (len(outliers) / len(df_clean)) * 100

        if len(outliers) > 0:
            # Crear columna indicadora de outliers
            df_clean[f'{col}_outlier'] = ((df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)).astype(int)
            outliers_report.append({
                'Columna': col,
                'Outliers': len(outliers),
                'Porcentaje': outlier_percent,
                'Límite inferior': lower_bound,
                'Límite superior': upper_bound
            })

if outliers_report:
    print("\nOutliers identificados (se han creado columnas indicadoras):")
    print(pd.DataFrame(outliers_report))
else:
    print("No se identificaron outliers significativos")
```

## **17. Transformación de Variables**

```python
print("\n=== TRANSFORMACIÓN DE VARIABLES ===")

# 1. Convertir SeniorCitizen a categórico
if 'SeniorCitizen' in df_clean.columns:
    df_clean['SeniorCitizen'] = df_clean['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
    print("'SeniorCitizen' convertido a categórico")

# 2. Crear variable binaria para Churn
if 'Churn' in df_clean.columns:
    df_clean['Churn_binary'] = (df_clean['Churn'] == 'Yes').astype(int)
    print("Variable objetivo 'Churn_binary' creada (1=Yes, 0=No)")

# 3. Crear banderas para servicios adicionales
service_cols = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
               'TechSupport', 'StreamingTV', 'StreamingMovies']

for col in service_cols:
    if col in df_clean.columns:
        df_clean[f'{col}_flag'] = (df_clean[col] == 'Yes').astype(int)
        print(f"Variable binaria '{col}_flag' creada")

# 4. Crear categorías de tenure
if 'tenure' in df_clean.columns:
    bins = [0, 12, 24, 36, 48, 60, 72]
    labels = ['0-1 año', '1-2 años', '2-3 años', '3-4 años', '4-5 años', '5+ años']
    df_clean['tenure_group'] = pd.cut(df_clean['tenure'], bins=bins, labels=labels, right=False)
    print("Variable categórica 'tenure_group' creada")
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

# Suponiendo que df_clean es su DataFrame después de los pasos de limpieza

# Preparar las características (X) y el objetivo (y)
# X Contiene columnas numéricas y categóricas. 'Contract'
X = df_clean[['tenure', 'Charges.Monthly', 'Contract']]

# Utilice la columna 'Churn' limpiada para la variable de destino
y = df_clean['Churn_clean'] # Utilice la columna que se limpió específicamente para NaN

# Verifique si hay NaN en la variable de destino antes de continuar
if y.isnull().any():
    print("Warning: NaN valores que aún se encuentran en la variable objetivo 'Churn_clean'.")
    # Es posible que desee investigar por qué los NaN permanecen o eliminan filas con NaN en y
    # Para este ejemplo, eliminaremos filas con NaN en y
    original_rows = len(X)
    combined_df = pd.concat([X, y], axis=1)
    combined_df.dropna(subset=['Churn_clean'], inplace=True)
    X = combined_df.drop(columns=['Churn_clean'])
    y = combined_df['Churn_clean']
    print(f"Abandonó {original_rows - len(X)} filas con NaN en 'Churn_clean'.")


# Codificación one-hot de la columna 'Contrato'
X = pd.get_dummies(X, columns=['Contract'], drop_first=True) # drop_first=True evita la multicolinealidad

# Comprueba la nueva estructura de X
print("Estructura de X después de la codificación one-hot:")
print(X.head())
print("\nTipos de datos en X:")
print(X.dtypes)

# Asegúrese que y tenga el formato correcto (por ejemplo, numérico si es necesario, RandomForest maneja cadenas)
# RandomForestClassifier Puede manejar etiquetas de cadena para la clasificación, garantiza la consistencia.
# Si 'Yes'/'No' son los valores, debería funcionar.

# Ahora, ajuste el modelo con los datos procesados.
model = RandomForestClassifier()
model.fit(X, y)

print(f"Precisión del modelo: {model.score(X, y):.0%}")
```

# 🔍 **Análisis de Correlación para Identificar Drivers de Churn**

## **Preparación de Datos para Análisis de Correlación**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Copia del dataframe para análisis
df_corr = df_clean.copy()

# Convertir variables categóricas a numéricas
le = LabelEncoder()
# Incluye 'Churn_clean' en la lista de columnas categóricas si está presente
categorical_cols = ['gender', 'Partner', 'Dependents', 'PhoneService',
                   'PaperlessBilling', 'Churn']
# Asegúrese de que las columnas existan antes de intentar codificarlas
categorical_cols_to_encode = [col for col in categorical_cols if col in df_corr.columns]
for col in categorical_cols_to_encode:
    # Asegúrese de que la columna no contenga tipos de datos complejos si el codificador de etiquetas no puede manejarlos
    if df_corr[col].dtype == 'object':
        # Reemplace los valores que no sean string o diccionarios con un marcador de posición si es necesario,
        # o elimine los valores nulos, dependiendo de cómo desee manejar esto
        # Para este ejemplo, simplemente eliminaremos los valores nulos para la codificación.
        non_null_values = df_corr[col].dropna()
        if not non_null_values.empty:
             df_corr[col] = le.fit_transform(df_corr[col].astype(str)) # Convertir a string para evitar errores con tipos mixtos
        else:
             print(f"Warning: La columna '{col}' está vacía o solo contiene valores nulos. Saltando la codificación.")
    else:
        print(f"Warning: La columna '{col}' no es de tipo 'object'. Saltando la codificación LabelEncoder.")


# Crear variable de conteo de servicios adicionales
services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
            'TechSupport', 'StreamingTV', 'StreamingMovies']

# Filtrar la lista de servicios para incluir solo las columnas presentes en df_corr
available_services = [svc for svc in services if svc in df_corr.columns]

# Verificar si hay servicios disponibles para contar
if available_services:
    # Utiliza solo las columnas disponibles para el cálculo
    df_corr['TotalServices'] = df_corr[available_services].apply(lambda x: (x == 'Yes').sum(), axis=1)
    print("Columna 'TotalServices' creada con éxito.")
else:
    print("No se encontraron columnas de servicio ('OnlineSecurity', 'OnlineBackup', etc.) en el DataFrame.")
    # Opcionalmente, puede crear la columna con valores predeterminados (ej: 0) si no se encuentran servicios
    # df_corr['TotalServices'] = 0


# Calcular correlaciones
# Asegúrese de que haya suficientes columnas numéricas y no todas nulas antes de calcular la correlación.
# Además, use la columna numérica 'Churn_binary' si existe, en lugar de 'Churn'.
columns_for_correlation = df_corr.select_dtypes(include=np.number).columns.tolist()

if 'Churn_binary' in df_corr.columns:
    # Utilizar 'Churn_binary' para la correlación si está disponible
    if 'Churn' in columns_for_correlation:
         columns_for_correlation.remove('Churn') # Eliminar la columna original 'Churn' si existe y es numérica
    if 'Churn_binary' not in columns_for_correlation:
         columns_for_correlation.append('Churn_binary') # Asegurar que 'Churn_binary' esté incluida

# Filtrar columnas con solo un valor único o todos nulos, ya que la correlación será NaN
processable_corr_cols = [col for col in columns_for_correlation if df_corr[col].nunique() > 1 and not df_corr[col].isnull().all()]

if len(processable_corr_cols) > 1: # Necesita al menos dos columnas para calcular la correlación
    correlation_matrix = df_corr[processable_corr_cols].corr(numeric_only=True)
    print("\nMatriz de correlación calculada.")

    # Visualización de la matriz de correlación
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Matriz de Correlación de Variables Numéricas')
    plt.show()
else:
    print("\nNo hay suficientes columnas numéricas procesables para calcular la matriz de correlación.")
```
